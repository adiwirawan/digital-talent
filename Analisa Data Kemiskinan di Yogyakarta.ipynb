# [1] Deskripsi Data
Analisis dilakukan dengan melihat data kemiskinan di Yogyakarta pada tahun 2007 sampai dengan tahun 2013

```python
import pandas as pd
master_data= pd.read_csv('processed-data-blm-kemiskinan-di-yogyakarta-2007-2013.csv', delimiter = ',', encoding = "utf-8")
master_data
```

# [2] Statistik Data
Statistik data mengetengahkan beberapa data yang bisa dilihat dengan data yang ada di bawah ini.

```python
print("Rata-rata penduduk miskin di Yogya dari 2007-2013 = ",master_data.jumlah_penduduk_miskin.mean())
print("Maximum jumlah penduduk miskin di Yogya dari 2007-2013 = ",master_data.jumlah_penduduk_miskin.max())
print("Minimum jumlah penduduk miskin di Yogya dari 2007-2013 = ",master_data.jumlah_penduduk_miskin.min())
```

# [2.1] Kecamatan Dengan Jumlah Kemiskinan Terbesar
Pada rentang tahun 2007 sampai dengan 2013 Kecamatan Gedang Sari merupakan kecamatan dengan Jumlah Penduduk Miskin terbesar.

```python
master_data[master_data['jumlah_penduduk_miskin'] == master_data.jumlah_penduduk_miskin.max()]
```

# [2.2] Kecamatan Dengan Jumlah Kemiskinan Terkecil
Pada rentang tahun 2007 sampai dengan 2013 Kecamatan Pakualaman merupakan kecamatan dengan Jumlah Penduduk Miskin terbesar.

```python
master_data[master_data['jumlah_penduduk_miskin'] == master_data.jumlah_penduduk_miskin.min()]
```

```python
master_data[(master_data['kategori']=='Miskin')].groupby('nama_kabkota').mean()['jumlah_penduduk_miskin']
```

# [3] Visualisasi Sederhana
Ditampilkan pada tahun 2009. Dari keseluruhan Kabupaten yang ada di DIY, maka Kabupaten Gunung Kidul merupakan Kabupaten dengan jumlah angka kemiskinan terbesar.

```python
ax = master_data[(master_data['tahun'] == 2009)].groupby('nama_kabkota').mean()['jumlah_penduduk_miskin']
ax = ax.plot(kind='bar', title ="Kondisi Kemiskinan Provinsi DIY di Tahun 2009", legend=True, fontsize=20, figsize= (30,10))
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Rerata Jumlah Penduduk Miskin", fontsize=12)
```

Sedangkan pada tahun 2013, berdasar data yang ada, maka Bantul adalah kabupaten dengan jumlah penduduk miskin terbesar

```python
ax = master_data[(master_data['tahun'] == 2013)].groupby('nama_kabkota').mean()['jumlah_penduduk_miskin']
ax = ax.plot(kind='bar', title ="Kondisi Kemiskinan Provinsi DIY di Tahun 2013", legend=True, fontsize=20, figsize= (30,10))
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Rerata Jumlah Penduduk Miskin", fontsize=12)
```
